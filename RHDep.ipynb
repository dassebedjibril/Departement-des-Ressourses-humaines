### `KDE` (Kernel Density Estimte) est utilise pour evaluer la densite de probablite d une variable continue 
### `KDE` decrit la densite de probabilite a differente valeur dans une variable continue

##### COMPRENDRE L'INTUITION DERRIÈRE `LOGISTIC REGRESSION CLASSIFIERS` , `LES RÉSEAUX NEURONEAUX ARTIFICIELS` ET LE `RANDOM FOREST CLASSIFIER`

#### ENTRAINER ET ÉVALUER `LOGISTIC REGRESSION`

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

#### Tester la Performance


```python
from sklearn.metrics import confusion_matrix, classification_report
print('Accuracy {} %'.format(100 * accuracy_score(y_pred, Y_test)))
```

```python
cm = confusion_matrix(y_pred, Y_test)
sns.heatmap(cm, annot=True)
```

#### Nous avons correctement classifier 
##### 300 vrais positifs 
##### 15 vrais negatifs
##### 7 faux negatifs
##### 47 faux positifs

```python
print(classification_report(y_pred, Y_test))
```

#### ENTRAINER ET ÉVALUER `DEEP LEARNING`

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500, activation= 'relu', input_shape = (50, )))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer= 'Adam', loss='binary_crossentropy', metrics= ['accuracy'])
```

```python
epoch_hist = model.fit(X_train, Y_train, epochs=100, batch_size=50)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
plt.plot(epoch_hist.history['loss'])
plt.title("model Loss Progress during training")
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend('Training Loss')
```

```python
plt.plot(epoch_hist.history['accuracy'])
plt.title("model accurary Progress during training")
plt.xlabel('Epoch')
plt.ylabel('Training accuracy')
plt.legend('Training accurary')
```

```python
cm = confusion_matrix(Y_test, y_pred)
sns.heatmap(cm, annot=True)
```

#### Nous avons correctement classifier 
##### 290 vrais positifs 
##### 18 vrais negatifs
##### 23 faux positifs
##### 41 faux negatifs

```python
print(classification_report(y_pred, Y_test))
```
